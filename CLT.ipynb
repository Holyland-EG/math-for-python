```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

# ЦПТ для экспоненциального распределения (Python 3)

```python
#Выбрал экспоненциальное распределение
expon_rv = sts.expon(loc = 0, scale = 3.)
t = expon_rv.rvs(size = 1000) #сгенерировал выборку объёма 1000
plt.hist(t, density = True, label = 'sample hist') #строим гистограмму выборки

x = np.linspace(0,20,100)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical PDF') #и поверх неё теоретическую плотность распределения
plt.ylabel('$exp(-x/3)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
#Строим гистограмму выборочных средних для выборок размера n = 5
e_1 = []
for i in range(999):
    expon_rv = sts.expon(loc=0, scale=3.)
    t = expon_rv.rvs(size=5)
    e_1.append(np.mean(t))

plt.hist(e_1, density=True, label = 'sample average for n=5')
#И кривую плотности вероятности соотв. нормального распределения
x = np.linspace(0,10,100)
norm_rv = sts.norm(loc = 3, scale = 1./(3))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical PDF')
plt.ylabel('$Norm.dist.$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
#Строим гистограмму выборочных средних для выборок размера n = 50
e_2 = []
for i in range(999):
    expon_rv = sts.expon(loc=0, scale=3.)
    t = expon_rv.rvs(size=50)
    e_2.append(np.mean(t))

plt.hist(e_2, density=True, label = 'sample average for n=50')
#И кривую плотности вероятности соотв. нормального распределения
x = np.linspace(0,10,100)
norm_rv = sts.norm(loc=3, scale=1./(3))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical PDF')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
#Строим гистограмму выборочных средних для выборок размера n = 100
e_3 = []   
for i in range(999):
    expon_rv = sts.expon(loc=0, scale=3.)
    t = expon_rv.rvs(size=100)
    e_3.append(np.mean(t))

plt.hist(e_3, density=True, label = 'sample average for n=100')
#И кривую плотности вероятности соотв. нормального распределения
x = np.linspace(0,10,100)
norm_rv = sts.norm(loc=3, scale=1./(3))
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label = 'theoretical PDF')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

Выводы: с увеличением n гистограммы выборычных средних приближаются к графику нормального распределения с соответствующими теоретическими средним и дисперсией
