```python
import numpy as np
import scipy.stats as sts
import math

import matplotlib.pyplot as plt
%matplotlib inline
```

# ЦПТ для распределения Лапласа (Python 3)

```python
laplace = sts.laplace()

#генерация выборки
matrix = laplace.rvs(1000)
matrix
```

```python
plt.hist(matrix, normed=True, label = 'sample hist')

x = np.linspace(-7,7,100)

pdf = laplace.pdf(x)
plt.plot(x, pdf, 'o', label = 'theoretical PDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

```python
# Дисперсия
D = laplace.var()

# Мат. ожидание
M = laplace.mean()
```

```python
#Строим гистограмму выборочных средних для выборок размера n = 5,10,50,100
for n in [5,10,50,100]:
    l = []
    for i in range(len(matrix)):
        laplace = sts.laplace()
        sample = laplace.rvs(size=n)
        l.append(np.mean(sample))

    plt.hist(l, density=True, label = 'sample average for n={}'.format(n))
    
    
   
    x = np.linspace(-2,2,100)
    norm_rv = sts.norm(loc=M, scale=np.sqrt(D/n))
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label = 'theoretical PDF')
    
    plt.ylabel('$Norm.dist.$')
    plt.xlabel('$x$')
    plt.legend(loc='upper right')
    plt.show()
```
